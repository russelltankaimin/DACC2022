```python
df = pd.read_csv("/kaggle/input/dacc-jk/JK_Part0.csv")
for i in range(1,10):
    df1 = pd.read_csv("/kaggle/input/dacc-jk/JK_Part"+str(i)+".csv")
    df = pd.concat([df,df1])
df
```

```python
df["osname"].value_counts()
```

```python
dff_ios = df[df["osname"] == "ios"].sample(n=500000,random_state = 100)
dff_android = df[df["osname"] == "android"].sample(n=500000,random_state = 100)
dff = pd.concat([dff_ios,dff_android])
dff
```

```python
dff["osname"] = dff["osname"].replace(["ios","android"],[1,-1])
dff["driving_mode"] = dff["driving_mode"].replace(["car","motorcycle"],[1,-1])
dff["osname"]
```

```python
print(dff["osname"].value_counts())
```

```python
dff_y = dff["accuracy"]
dff = dff.drop(columns=["accuracy","trj_id","pingtimestamp"])
dff_y
```

```python
model = LinearRegression()
model.fit(dff,dff_y)
importance = model.coef_
for i,v in zip(dff.columns,importance):
	print('Feature: %s, Score: %.5f' % (i,v))
# plot feature importance
plt.bar([x for x in dff.columns], importance)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.inspection import permutation_importance
model = KNeighborsRegressor()
# fit the model
model.fit(dff,dff_y)
# perform permutation importance
results = permutation_importance(model, dff,dff_y, scoring='neg_mean_squared_error')
# get importance
importance = results.importances_mean
# summarize feature importance
for i,v in zip(dff.columns,importance):
	print('Feature: %s, Score: %.5f' % (i,v))
# plot feature importance
plt.bar([x for x in dff.columns], importance)
plt.show()
```
